# **Supply Chain is the network of production and logistics involved in producing and delivering goods to customers.**

# **Supply Chain to understand how to imporve the effectiveness of the  Supply Chain to create more value for customers.**

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go

data=pd.read_csv("/content/supply_chain_data.csv")

data.head(10)
```

# **Let's have a look at the descriptive statics of the dataset:**

```python
data.describe()
```

# **Looking at the relationship between the price of the products and the revenue generated by them:**

```python
fig=px.scatter(data,x="Price",
               y="Revenue generated",
               color="Product type",
               hover_data=["Number of products sold"],
               trendline="ols")
fig.show()
```

# **Let's have a look at the sales by Product Type:**

```python
sales_data=data.groupby("Product type")["Number of products sold"].sum().reset_index()

pie_chart= px.pie(sales_data,values="Number of products sold",names="Product type",
                  title="Sales by Product Type",hover_data=["Number of products sold"],hole=0.5,
                  color_discrete_sequence=px.colors.qualitative.Pastel)

pie_chart.update_traces(textposition="inside",textinfo="percent+label")

pie_chart.show()
```

```python
data
```

# **Let's have a look at the total revenue generated from shipping carriers:**

```python
total_revenue=data.groupby("Shipping carriers")["Revenue generated"].sum().reset_index()
fig=go.Figure()

fig.add_trace(go.Bar(x=total_revenue["Shipping carriers"],
                     y=total_revenue["Revenue generated"]))

fig.update_layout(title="Total Revenue by Shipping Carrier",
                  xaxis_title="Shipping Carrier",
                  yaxis_title="Revenue Generated")

fig.show()
```

# **Let's have a look at the average lead time and average manufacturing costs for all products of the company:**

```python
avg_lead_time=data.groupby("Product type")["Lead time"].mean().reset_index()

avg_manufacturing_costs = data.groupby("Product type")["Manufacturing costs"].mean().reset_index()
result=pd.merge(avg_lead_time,avg_manufacturing_costs,on="Product type")

result.rename(columns={"Lead time":"Average Lead Time","Manufacturing costs":"Average Manufacturing costs"})

result
```

SKU = Stock Keeping Units
You have a large toy store with lots of toys. Each toy is different and has its name and price, but when you want to know how many have left, you need to identify them. So you can give each toy a unique code like a secret number only the the store knows. This secret number is actually called SKU. 

# **Let's Analyze the revenue generated by each SKU (SKU = product):**

```python
revenue_chart=px.line(data,x="SKU",y="Revenue generated",title="Revenue Generated by SKU")

revenue_chart.show()
```

# **Let's have a look at the stock levels of each SKU:**

# **Cost Analysis**

# **Let's analyze the shipping cost of Carriers:**

```python
shipping_cost_chart=px.bar(data,x="Shipping carriers",
                           y="Shipping costs",
                           title="Shipping Costs by Carrier")

shipping_cost_chart.show()
```

# **Let's have alook at the cost distribution by transportation by mode:**

```python
transporatation_chart=px.pie(data,values="Costs",names="Transportation modes",title="Cost Distribution by Transportation Mode",
                             hole=0.5,
                             color_discrete_sequence=px.colors.qualitative.Pastel)

transporatation_chart.show()
#this is how much they spent on transporation
```

# **Analyzing Defect Rate**

# **Let's have a look at the average defect rate of all product types:**

```python
defect_rates_by_product=data.groupby("Product type")["Defect rates"].mean().reset_index()

fig=px.bar(defect_rates_by_product,x="Product type",y="Defect rates",title="Average Defect Rates By Product Type")

fig.show()
```

# **Let's have damage rates by the mode of transporation:**

```python
pivot_table=pd.pivot_table(data,values="Defect rates",index=["Transportation modes"],aggfunc="mean")

transportation_chart=px.pie(values=pivot_table["Defect rates"],names=pivot_table.index,title="Defect Rates by Transportation Mode",
                            hole=0.5,
                            color_discrete_sequence=px.colors.qualitative.Pastel)
transportation_chart.show()
```
